```python
#@title Imports

import requests
from bs4 import BeautifulSoup
```

```python
#@title Extracting All the links and TAGS from A Website

#Define the seed url
seed_url = 'https://www.dawn.com/'

#Make a request to seed url
response = requests.get(seed_url)

#Parse TML response
soup = BeautifulSoup(response.content, 'html.parser')

#FInd all of the links on the page
links = soup.find_all('a')

#Print all the links on website
for link in links:
  print(link.get('href'))
```

```python
#@title Extracting Important Headins from A Website

#Define the seed url
seed_url = 'https://www.dawn.com/'

#Make a request to seed url
response = requests.get(seed_url)

#Parse TML response
soup = BeautifulSoup(response.content, 'html.parser')

#FInd all of the h2Tags on the page
h2_tags = soup.find_all('h2')

#Extract the text from H2Tags
h2_texts = [h2.text for h2 in h2_tags]

#Remove Empty Strings From List
h2_texts = [text for text in h2_texts if text]

# Remove duplicates
h2_texts = list(set(h2_texts))

#Print the extracted text
print('===========Major Topics Being Discussed On This Website===========\n')
for text in h2_texts:
  print(text)
```

```python
#@title Extract All The image URL's from A Website

# Define the seed URL
seed_url = 'https://www.dawn.com/'

# Request and parse the HTML
response = requests.get(seed_url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all <img> tags
img_tags = soup.find_all('img')

# Extract image URLs (check both 'src' and 'data-src' for lazy loading)
img_urls = []
for img in img_tags:
    src = img.get('src') or img.get('data-src')
    if src:
        img_urls.append(src)

# Remove duplicates
img_urls = list(set(img_urls))

# Print the image URLs
print('=========== Image URLs ===========\n')
for url in img_urls:
    print(url)
```

```python
#@title Extract All The Video URL's from A Website

# Define the seed URL
seed_url = 'https://www.dawn.com/'

# Request and parse the HTML
response = requests.get(seed_url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all possible video-related tags
video_tags = soup.find_all(['video', 'source', 'iframe'])

# Extract video URLs
video_urls = []
for tag in video_tags:
    src = tag.get('src') or tag.get('data-src')
    if src and any(ext in src for ext in ['.mp4', 'youtube.com/', 'video']):
        video_urls.append(src)

# Remove duplicates
video_urls = list(set(video_urls))

# Print the video URLs
print('=========== Video URLs ===========\n')
for url in video_urls:
    print(url)
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
```

```python
#traverse mutiple web pages
visited= set()
def traverse_mutiple_pages(url, domain, depth=0, max_depth=2):
    if url in visited or depth>max_depth:
        return
    response = requests.get(url)
    visited.add(url)
    print(f"Visited this link: {url}")
    soup = BeautifulSoup(response.content, 'html.parser')
    for link in soup.find_all('a',href=True):
        href= link['href']
        next_url=urljoin(url,href)

        if urlparse(next_url).netloc == domain:
            traverse_mutiple_pages(next_url,domain,depth+1,max_depth )



seed_url = 'https://www.dawn.com/'
domain= urlparse(seed_url).netloc
traverse_mutiple_pages(seed_url,domain)
    
```

```python
import json
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse, urljoin

from collections import defaultdict

def normalize_base_url(url):
    """Remove query params and fragments — keep scheme + netloc + path only."""
    parsed = urlparse(url)
    return f"{parsed.scheme}://{parsed.netloc}{parsed.path.rstrip('/')}"

def store_structured_data(visited):
    grouped_data = defaultdict(lambda: {
        "url": "",
        "headings": set(),
        "links": set(),
        "images": set()
    })

    for current_url in visited:
        try:
            response = requests.get(current_url, timeout=5)
            soup = BeautifulSoup(response.content, 'html.parser')

            base_url = normalize_base_url(current_url)
            grouped_data[base_url]["url"] = base_url

            h2_tags = soup.find_all('h2')
            for h2 in h2_tags:
                text = h2.text.strip()
                if text:
                    grouped_data[base_url]["headings"].add(text)

            links = soup.find_all('a', href=True)
            for link in links:
                href = link['href']
                full_link = urljoin(current_url, href)
                if full_link.startswith(base_url):  
                    grouped_data[base_url]["links"].add(full_link)

            img_tags = soup.find_all('img')
            for img in img_tags:
                src = img.get('src') or img.get('data-src')
                if src:
                    full_img = urljoin(current_url, src)
                    grouped_data[base_url]["images"].add(full_img)

        except Exception as e:
            print(f"Failed to process {current_url}: {e}")

    structured_data = []
    for page in grouped_data.values():
        structured_data.append({
            "url": page["url"],
            "headings": list(page["headings"]),
            "links": list(page["links"]),
            "images": list(page["images"]),
        })

    with open('grouped_data.json', 'w', encoding='utf-8') as f:
        json.dump(structured_data, f, ensure_ascii=False, indent=2)


store_structured_data(visited)
```
